# 1. Nested Lists (or 2D Lists)

#### * Just like we have loops within loops, we can have lists within lists.  

#### * A "1D" list contains numbers; a "2D" list contains lists -- like a table!



```python
# EXAMPLE 1a: A list of lists -- or, a "2-dimensional" list.

my_table = [[51,42,36,81],[41,11,52,63],[91,81,32,10,77]]

print('len(my_table) =', len(my_table))
print('my_table[1] =', my_table[1])
print('my_table[1][3] =', my_table[1][3])

my_table[0][2] += 1000
print('my_table =', my_table)
```

#### * Notice:

#### --- The lists contained in a list don't have to be the same length. 

#### ---  We access (and update) entries using two pairs of brackets.   E.g. `my_table[1]` is the second element of `my_table`, so `(my_table[1])[3]` just means the fourth element of the list `my_table[1]`.  

#### ---  `len(my_table)` is the number of elements the list contains -- what those elements are has no bearing on this calculation.

# 2. Pretty Printing and Row-Summing

#### * How can we pretty print a 2D list? Nested `for`-loops!

```python
# EXAMPLE 2a: Pretty print a 2D List

my_table = [[51,42,36,81],[41,11,52,63],[91,81,32,10,77]]

# LET'S PRINT PRETTY!
# for row in my_table:
#     for column in row:
#         print(f'{column:4}', end = '')
#     print()

for y in range(len(my_table)):
    for x in range(len(my_table[y])):
        print(f'{my_table[y][x]:4}',end = '')
    print()
        
```

#### * Print out the sum of all the entries in each row (without using the sum function).  

```python
# EXAMPLE 2b: Print out the sum of each row of my_table.

my_table = [[51,42,36,81],[41,11,52,63],[91,81,32,10,77]]

# WRITE NESTED FOR LOOPS WHICH CALCULATE THE THREE SUMS
# CORRECT ANSWERS: 210, 167, 291.
for row in my_table:
    s = 0
    for col in row:
        s += col
    print(s)
```

# 3. Updating 2D Lists, Looping By Index 

#### * Let's take `my_table`, and multiply each entry by 10.  How do we do that?  

#### * Certainly not by `my_table = my_table*10`.

#### * Need a nested loop, of course!  

#### * BUT: if you want to update, loop through *positions*, not *values*. So there are going to be some `range(len())`'s here.

```python
# EXAMPLE 3a: Update a 2D List

my_table = [[51,42,36,81],[41,11,52,63],[91,81,32,10,77]]

NUM_ROWS = len(my_table)
for row_idx in range(NUM_ROWS):           
    
    for col_idx in range(len(my_table[row_idx])):    
        my_table[row_idx][col_idx] *= 10
        
print(my_table)
```

# 6. Cat Transmogrification

#### * We can change images by adjusting pixels.  Watch this.

```python
# EXAMPLE 6a: Scary Cat

from skimage import io # Import for reading and converting images
import matplotlib.pyplot as plt # Import for displaying images

cat = io.imread('cat.jpg') # Function for loading


for row_idx in range(400):
    for col_idx in range(400):
        # For pixel (row_idx, col_idx), this code "flips" all the levels.  If a pixel has a level near 0, that level will now be near
        # 255, and vice-versa. What effect does this have?
        cat[row_idx][col_idx][0] = 255-cat[row_idx][col_idx][0]
        cat[row_idx][col_idx][1] = 255-cat[row_idx][col_idx][1]
        cat[row_idx][col_idx][2] = 255-cat[row_idx][col_idx][2]

# These two lines prepare a display, and then actually display it.
plt.imshow(cat)
plt.show()
```

<br><br><br><br><br>
<br><br><br><br><br>

#### * Let's boost red-levels, using the following rules: 

#### --- if the red level of a pixel is less than 128, we'll multiply by 2 (we don't want the red level going over 255!)

#### --- if the red level of a pixel is greater than or equal to 128, we'll just set the red level to be 255.


#### * If that's too easy: superimpose the French flag on the cat. (A vertical blue stripe, vertical white strip, vertical red stripe.)

```python
# EXAMPLE 6b: Red Cat

from skimage import io # Import for reading and converting images
import matplotlib.pyplot as plt # Import for displaying images

cat = io.imread('cat.jpg') # Function for loading

# There are 400 rows, each with 400 pixels.
# For each pixel, change the red value as described above.  
# Hint: the red value is cat[row_idx][col_idx][0].

for row_idx in range(400):
    for col_idx in range(400):
        # For pixel (row_idx, col_idx), this code "flips" all the levels.  If a pixel has a level near 0, that level will now be near
        # 255, and vice-versa. What effect does this have?
         
        if cat[row_idx][col_idx][0] < 128:
            cat[row_idx][col_idx][0] *= 2
        else:
            cat[row_idx][col_idx][0] = 255
#         cat[row_idx][col_idx][1] = 255-cat[row_idx][col_idx][1]
#         cat[row_idx][col_idx][2] = 255-cat[row_idx][col_idx][2]






plt.imshow(cat)
plt.show()
```

# 7. A Deeply Disturbing Phenomenon

#### * ?!?!

```python
# EXAMPLE 7a: Assignment

# Compare the results of this bit of code....
x = 'Old'
y = x
x = 'New'
print(y)

# ...with this one... #
x = ['Old', 'One']    #
y = x                 #   BLOCK A
x = ['New', 'Guy']    #
print(y)              #

#...AND THIS ONE!!!!! #
x = ['Old', 'One']    #
y = x                 #   BLOCK B
x[0] = 'New!!!!'      #
print(y)              #
```

#### * Why are the last two snippets so different?  They both change `x`, but one change causes `y` to change, and the other doesn't. 


#### * The short answer is:

#### --- the line `y = x` (in Block A and Block B), makes `y` and `x` refer to the same object in memory;

#### --- the subsequent line in Block A, starting with `x = `, reassigns `x` to an *entirely new object*, while `y` still is bound to the old object;

#### --- the subsequent line in Block B, starting with `x[0] = `, is a *mutation* which alters, but does not replace, the object which `x` and `y` both point to.

# 8. The Truth About Variables and Objects



#### * If I ask you to multiply 5201 by 3157, you probably can't do that in your head: you need to write down the results of intermediate computations, on a piece of paper. 

#### * If I ask the computer to multiply 5201 by 3157, it needs to "write things down", too -- in the  Random Access Memory, or RAM (we'll just call it the memory).  

#### * We use our eyes to find the things we've written down previously.  The computer uses numbers called *memory addresses* to keep track of where it exactly it wrote everything down.

#### * I'll use `0x` to indicate that I'm writing down a memory address in the examples I show you.

---------

#### * For every value Python needs -- every literal value it encounters, and every evaluation of an expression -- it creates an *object*.

#### * An **object** is a parcel of memory which contains a *value* and information about what *data type* that value represents, and which has an *address*.   

#### * Python variables don't really have values; they have objects that they are attached to, and those objects contain values.  

#### * Python maintains a sort-of "table", which matches defined variables to the addresses of the  objects they are bound to.

# 11. Solving the Riddle

```python
# EXAMPLE 11a: The weird code.

x = ['Old', 'One']  # Line 1a
y = x                # Line 2a
x = ['New', 'Guy']  # Line 3a

x = ['Old', 'One']  # Line 1b
y = x                # Line 2b
x[0] = 'New!!!!'     # Line 3b
```

#### * Line 1a/b: Creates objects for `"Old"`, `"One"`, and the list; puts the addresses of the strings into the list; and binds `x` to the list object.


#### * Line 2a/b: `x` is bound to the object at `0x23`, so this object now gets bound to `y` as well.

#### * *The difference:*

#### --- Line 3a: Creates objects for `"New"`, `"Guy"`, and a new list object to hold them; puts the addresses of the strings into the list; and binds `x` to the new list object.  **The old object at `0x23` isn't touched, and `y` is still bound it it.**

#### --- Line 3b: Modifies the contents of the list object at `0x23`.  Note that neither of the variable bindings change -- they both still point to `0x23`, which has now been changed!
